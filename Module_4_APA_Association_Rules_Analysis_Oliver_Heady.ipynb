### Create Dummy Variables for all Order Items

```python
# drop rows with missing data (hint: we learned this in module 2)
df = df.dropna() # drop all rows with missing data

data = list(df['Order_Items'].apply(lambda x:x.split(","))) # convert the order_items column to a list
```

```python
from mlxtend.preprocessing import TransactionEncoder # import the TransactionEncoder function from mlxtend.preprocessing, this is a more efficient means of creating the dummy columns

a = TransactionEncoder() # assign the funciton to an object called 'a'

a_data = a.fit(data).transform(data) # create a new object where the 'order_items' list has been converted to dummy variable columns
df = pd.DataFrame(a_data, columns=a.columns_) # create a dataframe object called df from the new dummy variable columns
# view df
df
```

### Convert T/F to 1/0 and clean up dataframe

```python
# replace 'true' with a value of 1, and 'false' with a  value of 0
df.replace({False : 0, True : 1}, inplace = True)

# view df
df
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '') # remove duplicated columns by first trimming all leading and trailing whitespaces
df = df.loc[:,~df.columns.duplicated()] # drop duplicated columns
# view df, ensure all duplicated columns were dropped
df
```

**Conduct an association rules analysis with a minimum confidence level of 0.001. Change the support of your frequent itemsets to 0.01.**
